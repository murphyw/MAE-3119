# Experiment 1: Thermal Failure of Carbon Resistors

## 1.1 Introduction and Purpose

The purpose of this experiment is to observe the validity of Ohm's Law for resistors in common use in electronic circuitry. The effects of temperature on electrical resistance are measured by applying power to resistors at the manufacturer's rated power and beyond. By applying electrical power in both a constant voltage and a constant current mode, the feedback effect of thermal instability is observed. The experiment also serves as an aid in identifying the appearance of damaged resistors on circuit boards, and the possibility of mis-identifying color codes on thermally damaged resistors. The use of automated testing equipment illustrates the potential for many simple tests of this kind to be performed quickly. By pooling data from all experimenters in one laboratory, the statistics of large samples is illustrated, and the benefits (and pitfalls) of independent investigation are exposed.

## 1.2 Safety Precautions

This procedure has been reviewed by the Office of Environmental Health and Safety and found to be safe provided that the following precautions are taken:

1. The resistor tests should all be performed over the ceramic tiles provided to prevent fires. All combustible articles, such as note paper, should be kept well away from the test area. 

2. The tests should be performed with the students positioned at least three feet away from the resistor under test. 

3. The debris from testing should not be handled until the power is removed and sufficient time has passed that the material has cooled off. The debris should then be stored in the metal cans provided. 

4. Students with health problems or respiratory concerns should contact the lab assistant to perform an alternative exercise.

## 1.4 Procedure, Part I: Color Code Identification and Tolerance Statistics

1. Put the VOM into resistance-measuring mode and touch the two measurement leads together. The meter should read near zero. Write down this "residual resistance" measurement and subtract it from all measurements which follow.

2. Use one of the digital VOMs to measure the resistance of the high-power "standard" resistor. Pay special attention to the button labeled "Terminals", which must be set to "Front" to enable the terminals on the front of the meter. Write down the resistance value, which should be close to 22 Ω. 

3. Pass this resistor to the next group, which will repeat the measurement. Repeat this process until all twelve of the standards have been measured. 

4. Calculate a mean <x> and standard deviation σ for these twelve measurements.

5. Write your result for these two measurements on the chalkboard. When all of the groups have reported their values, write down the twelve mean values of resistance obtained by the twelve groups for the same batch of resistors. From these measurements, the instrumental uncertainty of the VOM is determined.

6. Unpack the twenty small resistors. Note the color code arrangement, which should read red-red-black-gold, starting with the color band nearest the end of the resistor. 

7. Measure the resistance of each of the twenty small resistors. If any of the units are significantly out of specification, discard them and request replacements. 

8. For twenty good resistors, calculate the mean and standard deviation of the resistance. Find the ratio of the standard deviation to the mean, and compare this value to the tolerance of the resistor.

## 1.5 Procedure, Part II: Power Supply Setup

1. Place the standard resistor on the ceramic tile to protect the lab benchtop from heat. 

2. Connect the programmable power supply to the high-power standard resistor, using the terminals marked "+25" and "COM". 

3. Set the VOM to measure DC voltage. Measure the voltage across the resistor with the VOM. 

4. Using the front-panel controls, set the power supply to deliver 10.00 volts, in voltage-limited mode. 

5. Run this test for five minutes. The resistor should become noticeably warm but not hot. 

6. At the end of this test, use the "Output On/Off" button to temporarily remove power from the resistor. 

7. Remove the resistor from the power source and immediately (before it cools off) measure its resistance using the VOM. Write down the warm resistance and set the standard aside. 

8. Ensure that the output terminals are disabled before continuing with part III.

## 1.6 Procedure, Part III: Resistor Tests

1. Without changing the voltage or current programming of the power supply, substitute one of the 22 Ω, 1/4 watt resistors for the high-power standard. Be sure that the resistor is located safely over the ceramic plate and keep your hands away from it. 

2. Apply power by pushing the "Output On/Off" button. The resistor should be destroyed within a few seconds. Make a sketch of the failed component in your notebook. 

3. Note the final resistance of the device. In some cases, the resistor will break apart into several pieces, in which case the final resistance is obviously infinity. In most cases, the carbon film, which is painted helically along the ceramic former, will rupture or neck down at some point along its length. In other cases, however, the resistor will become a short circuit (zero ohms), in which case the power supply will current limit to a safe value.

4. For more subtle failure modes, lower the power supplied to the device. For these tests, the current through the device as well as the voltage across the device will be monitored while the device is under test. Make a test setup using the circuit shown in Figure 1.1.

5. Re-attach the 50 watt standard and reset the power supply to provide sufficient voltage to produce 2.5 watts of DC power, i.e. 7.416 volts (=(22x2.5)^(1/2)). 

6. Remove power from the terminals using the "Output On/Off" button and then replace the standard with one of the 22 Ω quarter-watt resistors. 

7. Turn on the power supply output and observe the current. Write down each value of the current at five-second intervals until the resistor fails or two minutes have elapsed, whichever happens first.

8. Repeat the above tests but with the power supply in current-controlled mode. Set the power supply to deliver 0.337 ampere (=(2.5/22)^(1/2)) constant current, using the 50 watt resistor as the load. 

9. Turn off the output, replace the fifty watt resistor with another quarter watt resistor, and apply power. This time observe the voltage across the resistor, writing down the value every five seconds, as above. 

10. Compare the time to failure between the two tests. Write down the time to failure on the chalkboard for the two tests, so that the times to failure for both the voltage-mode and current-mode tests can be analyzed statistically for all experiments in the laboratory.

## 1.7 Procedure, Part IV: Computer-Controlled Tests

1. Get the computer workstation up and running HP VEE. Log in as a user (EE100USER) with a password (123456). Be sure that the domain is set to Instructional. 

2. Open the HPVEE program by double-clicking the icon. From the file menu, select the stored program "restest". 

3. Change the file names for data storage for each resistor test, so that there are unique names for each file in the series. 

4. Initiate each measurement session by clicking on the "Play" button (filled triangle). Write down the specific conditions of each test in your laboratory notebook. 

5. Perform a series of ten tests, at nominal power levels of 0.25, 0.5, 1, 2, and 4 watts, in both constant-current mode and constant-voltage mode. 

6. Record the dependent parameter (i.e. voltage for the constant current mode or current for the constant-voltage mode) using the preset limits of sixty measurements at one-second intervals. 

7. For the low-power tests, where the resistor is hardly damaged, let the resistor cool to room temperature after the test, and then reserve the resistor for a cold resistance check of each component after all of the computer-driven tests are over.

## 1.9 Additional Questions for Laboratory Report

1. With the addition of heat, some color bands can shift. Red often becomes brown. Did any of your resistors become modified by heat so that they had a "correct" resistance for a color code of brown-brown-black at some point? What resistance is this?

2. From your statistical study of the fresh quarter-watt resistors, how did the standard deviation, expressed as a percentage, compare to the stated tolerance of the resistors? What fraction of the batch of resistors would you expect to be out of tolerance? (Hint: use a Gaussian probability distribution.)

3. Why do resistors operating at constant current fail at different rates from resistors at constant voltage, for the same initial power level?

4. Suppose that the failure rate of resistors follows a law like MTBF = kP^(-m), where m is an adjustable exponent and P is the power. A resistor operating at its rated power has a mean time between failure (MTBF) of five years. What value of m best describes your results?

5. How do you think the mounting of the resistor on a printed circuit board affects its power handling ability? Would these resistors be safe on a combustible substrate, like polyethylene? What precautions would you take in using circuits with these resistors for instrumentation in an oil well?

6. Resistors are cheaper than fuses and are often used to protect a more valuable component, such as a large integrated circuit, from damage. What is the equivalent "fuse current rating" of the 22 Ω, 1/4 W resistors you studied, to open within ten seconds?
