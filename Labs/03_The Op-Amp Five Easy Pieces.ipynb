# Experiment 3: The Op-Amp: Five Easy Pieces

## 3.1 Introduction and Purpose

The purpose of this experiment is to introduce the most important analog building block known as the operational amplifier (op-amp). The op-amp's versatility is demonstrated through five simple circuit configurations which illustrate its capability of providing voltage gain, load isolation, frequency-dependent operation, and its application as an instrumentation amplifier providing rejection of interfering signals. The limitations on the op-amp's performance are also shown through some simple tests.

## 3.3 Procedure, Part I: Op-Amp Pinout Identification and Power Hookup

1. Identify Pin 1 of the op-amp IC (marked by a small circular indentation in the lower left-hand corner of the package when viewed from the top side).
2. Note the important pins: Inverting Input (Pin 2), Non-Inverting Input (Pin 3), Negative Supply (Pin 4), Output (Pin 6), and Positive Supply (Pin 7).
3. Insert the op-amp on the prototype board, straddling one of the grooves separating the columns of sockets.
4. Connect power to the prototype board with the op-amp inserted: positive rail to Pin 7 and negative rail to Pin 4.
5. Turn on the power supply and measure the voltage at pins 7 and 4 to ground (should be +15 and -15 volts respectively).
6. Monitor the temperature of the op-amp package. It should be only slightly warm to the touch.
7. Assemble the circuit shown in Figure 3.2, using a 10 kΩ resistor for R1 and a 100 kΩ resistor for R2.
8. Apply power and measure the output voltage (should be between -0.05 and +0.05 volts).

## 3.4 Procedure, Part II: Amplifier Circuit Tests

1. Non-inverting Amplifier (Figure 3.2):
   - Apply a 100 mV p-p, 1 kHz signal from the function generator.
   - Measure the RMS AC voltage at the input and output.
   - Display both input and output signals on the oscilloscope.
   - Determine the -3 dB point of the amplifier.

2. Inverting Amplifier (Figure 3.3):
   - Reconfigure the circuit and repeat measurements.
   - Observe the effect of increasing input amplitude to 10.0 V p-p.
   - Increase frequency to 50 kHz and observe the slew rate effect.

3. Non-inverting Buffer (Figure 3.4):
   - Configure the circuit and apply a 100 mV p-p, 1000 Hz signal.
   - Measure input and output signals.
   - Determine the -3 dB point.
   - Test the effect of adding a 51 Ω resistor to the input and output.

4. Integrator (Figure 3.5):
   - Configure the circuit using 10 kΩ for R1, 1.0 MΩ for R2, and 0.1 μF for C.
   - Apply a 100 mV p-p, 1 kHz sine wave.
   - Measure output at 100 Hz, 1 kHz, and 10 kHz.
   - Observe the response to a 1 V p-p, 1 kHz square wave input.

5. Differencing Amplifier (Figure 3.6):
   - Configure the circuit with R1 = 10 kΩ and R2 = 100 kΩ.
   - Apply a 100 mV p-p, 1.0 kHz signal differentially and then in common mode.
   - Calculate the Common Mode Rejection Ratio (CMRR).
   - Test the amplifier with a simulated EKG signal.
