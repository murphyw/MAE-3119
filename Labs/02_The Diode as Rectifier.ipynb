# Experiment 2: The Diode as Rectifier

## 2.1 Introduction and Purpose

The purpose of this experiment is to familiarize students with the properties of a simple nonlinear circuit and fundamental concepts of alternating current. This exercise also serves as the student's first exposure to the oscilloscope. By observing the waveforms present at various parts of the circuit, the performance of this powerful passive building block is demonstrated. The exercise also shows the utility of the waveform storage and analysis capabilities of HP VEE.

## 2.2 Equipment Required

The equipment supplied in the laboratory inc- RIGOL Technologies DG1022Z Function Generator, with associated computer interface

## 2.3 Procedure, Part I: Function Generator and Oscilloscope Setup

1. Using the nylon prototype board, assemble the circuit shown in Figure 2.1. Note the polarity marking for the diode. The banded end is the cathode, which connects to the ungrounded end of the resistor and capacitor.

2. Connect the function generator to the circuit and program it to deliver 10 volts peak-to-peak (p-p) at a frequency of 60 Hertz (Hz).

3. Set up the oscilloscope to view the waveform. Use 5 milliseconds per division on the x-axis timebase, auto trigger, and 2 volts per division on the y-axis.

4. Observe the sinusoidal waveform on the oscilloscope. It should have a period of 3.33 divisions (16.67 ms) and an amplitude of 5 divisions peak-to-peak (10 volts).

5. Move the oscilloscope input connection to the top of resistor R1 on the other side of the diode. Sketch the new waveform in your notebook.

6. Attach both DVMs to terminal 2 and ground. Make voltage measurements using "DC Volts" on the upper meter and "AC Volts" on the lower meter.

7. Remove the 0.1 μF capacitor and note the change in the waveform. Sketch this circuit in your notebook and record the new AC and DC voltages.

8. Replace the 0.1 μF capacitor with the 1 μF capacitor. Record the AC and DC voltages and sketch the waveform.

9. Use HP VEE to run the "rectifier-scope-chart" program and repeat the above tests. Print out copies of the oscilloscope display for your notebook.

10. Reconfigure the circuit to the full-wave rectifier shown in Figure 2.2. Repeat the measurements and HP VEE analysis for 0.1 μF capacitor, 1 μF capacitor, and no capacitor configurations.

## 2.4 Procedure, Part II: DC Voltage and Ripple Study

1. Using the full wave circuit with a 0.1 μF capacitor, measure the AC and DC voltages with the same 10 volt p-p input sine wave applied but with the frequency adjusted from 10 Hz to 10000 Hz.

2. Take measurements at the following frequency intervals:
   - 10 Hz increments below 100 Hz
   - 100 Hz increments from 100 Hz to 1000 Hz
   - 1000 Hz increments from 1000 Hz to 10 KHz

3. Use the HP VEE module called "FreqSweep" to automate this test if desired. Save your data to a floppy disk for later analysis.

## 2.6 Additional Questions for Laboratory Report

1. Using the half-wave rectifier data, construct a plot of voltage vs. time for all three measurements (C=0, 0.1, 1.0 μF) on one drawing. Identify similar and different parts of the waveform. Can you explain the waveform parts with a simple function like V(t)? What values of T would you use?

2. Examine your ripple voltage measurements in the full-wave circuit as a function of frequency for the 0.1 μF case. Compare the AC ripple at 600 Hz with the AC ripple measured at 60 Hz with the 1.0 μF capacitor. Consider the dimensionless parameter ωRC. What error is made in using the 0.1 μF, 600 Hz data for the 1.0 μF, 60 Hz case? Is this error within the capacitor tolerances used?

3. Aviation and military power supplies often use AC frequencies higher than the standard 60 Hz (e.g., 400 Hz for aircraft power). What is the benefit of higher frequency in terms of the size and weight of rectification circuitry?

4. The full-wave bridge circuit used here requires that the AC source be "floating" (neither side connected to earth ground). How would the circuit perform if one output terminal of the signal source were tied to the "ground" point in Figure 2.2?
