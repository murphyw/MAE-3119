# Experiment 6: Voltage Dividers and Wheatstone Bridge

## 6.1 Introduction and Purpose

The purpose of this experiment is to familiarize students with voltage divider circuits and the Wheatstone bridge, fundamental concepts in electrical engineering. These circuits are essential for various applications, including sensor calibration, resistance measurement, and signal conditioning. This lab will demonstrate the principles of voltage division, explore the behavior of loaded and unloaded voltage dividers, and utilize a Wheatstone bridge for precise resistance measurements.

## 6.3 Procedure, Part I: Simple Voltage Divider

1. Construct a voltage divider using a 10 kΩ and a 22 kΩ resistor on the prototype board.
2. Apply 10V DC from the power supply to the voltage divider.
3. Measure the output voltage of the divider using a digital voltmeter.
4. Calculate the expected output voltage and compare it with the measured value.
5. Repeat steps 1-4 using different resistor combinations provided.

## 6.4 Procedure, Part II: Loaded Voltage Divider

1. Using the 10 kΩ and 22 kΩ voltage divider, connect a 47 kΩ resistor in parallel with the 22 kΩ resistor to act as a load.
2. Measure the new output voltage and compare it with the unloaded case.
3. Calculate the loading effect and the new expected output voltage.
4. Repeat the experiment with a 10 kΩ load resistor.
5. Discuss the impact of loading on the voltage divider's output.

## 6.5 Procedure, Part III: Wheatstone Bridge

1. Construct a Wheatstone bridge circuit using three 10 kΩ resistors and one 10 kΩ potentiometer.
2. Apply 5V DC to the bridge circuit.
3. Adjust the potentiometer to balance the bridge (zero voltage across the middle).
4. Replace one 10 kΩ resistor with an unknown resistor.
5. Re-balance the bridge by adjusting the potentiometer.
6. Calculate the value of the unknown resistor based on the potentiometer setting.
7. Verify the calculated value using a digital multimeter.

## 6.6 Procedure, Part IV: AC Voltage Divider and Low-Pass Filter

1. Construct a voltage divider using a 10 kΩ resistor and a 0.1 μF capacitor.
2. Apply a 1 kHz, 5V peak-to-peak sine wave from the function generator.
3. Observe the output waveform on the oscilloscope.
4. Measure the output voltage amplitude and phase shift relative to the input.
5. Repeat the measurements for frequencies of 100 Hz, 500 Hz, 2 kHz, 5 kHz, and 10 kHz.
6. Plot the frequency response (gain in dB vs. frequency) of the circuit.

## 6.8 Additional Questions for Laboratory Report

1. How does the loading effect change if you use a voltage follower (buffer) circuit with an op-amp between the voltage divider and the load? Design and analyze such a circuit.

2. In the Wheatstone bridge experiment, how does the sensitivity of the measurement change if you increase the supply voltage to 10V? Explain the trade-offs involved in choosing the supply voltage for a Wheatstone bridge circuit.

3. For the AC voltage divider circuit, calculate the -3dB point theoretically and compare it to your measured results. How could you modify the circuit to change this cutoff frequency?

4. Design a voltage divider circuit that can be used as a simple temperature sensor using a thermistor. Provide a theoretical analysis of how the output voltage would change with temperature.

5. How would you modify the Wheatstone bridge circuit to measure small changes in capacitance instead of resistance? Describe the circuit modifications and the measurement procedure.
